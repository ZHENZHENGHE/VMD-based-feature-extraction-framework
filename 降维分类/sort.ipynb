```python
# -*- coding: utf-8 -*-
"""
08_publication_roc.py

Publication-ready ROC curves
"""

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import roc_curve, auc

# ============================================================
# PATH
# ============================================================

ROOT = Path(
    r"D:/a_work/课题组实验数据处理/新预处理/results/ml_validation_loso"
)

PRED_FILE = ROOT / "loso_ml_predictions.csv"

OUT_DIR = ROOT / "publication_figures"
OUT_DIR.mkdir(exist_ok=True)

# ============================================================
# LOAD
# ============================================================

df = pd.read_csv(PRED_FILE)

# ============================================================
# ROC
# ============================================================

plt.figure(figsize=(8, 7))

methods = ["event_guided", "fixed"]
colors = ["tab:blue", "tab:orange"]

for method, color in zip(methods, colors):

    # ------------------------------------------------
    # 只保留对应 method 的最佳模型
    # ------------------------------------------------

    if method == "event_guided":

        sub = df[
            (df["Method"] == "event_guided")
            &
            (df["Model"] == "RandomForest")
        ]

    elif method == "fixed":

        sub = df[
            (df["Method"] == "fixed")
            &
            (df["Model"] == "SVM_Linear")
        ]

    # ------------------------------------------------
    # ROC
    # ------------------------------------------------

    y_true = sub["TrueLabel"].values
    y_score = sub["Score"].values

    fpr, tpr, _ = roc_curve(y_true, y_score)

    roc_auc = auc(fpr, tpr)

    plt.plot(
        fpr,
        tpr,
        lw=3,
        color=color,
        label=f"{method} RF (AUC={roc_auc:.3f})"
    )

# random line
plt.plot([0, 1], [0, 1], "--", color="gray", lw=2)

plt.xlim([0, 1])
plt.ylim([0, 1.02])

plt.xlabel("False Positive Rate", fontsize=16)
plt.ylabel("True Positive Rate", fontsize=16)

plt.title(
    "LOSO ROC comparison: event-guided vs fixed-window",
    fontsize=18
)

plt.legend(fontsize=14)

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(
    OUT_DIR / "publication_roc_curve.png",
    dpi=600,
    bbox_inches="tight"
)

# plt.close()
plt.show()

print("ROC figure saved.")
```

```python
# -*- coding: utf-8 -*-
"""
09_recurrence_plot_examples.py

Recurrence plot visualization
"""

from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pyts.image import RecurrencePlot

# ============================================================
# PATH
# ============================================================

ROOT = Path(
    r"D:/a_work/课题组实验数据处理/新预处理/results"
)

OUT_DIR = ROOT / "ml_validation_loso" / "loso_ml_predictions"
OUT_DIR.mkdir(exist_ok=True)

# ============================================================
# SELECT SUBJECTS
# ============================================================

healthy_subject = "2bHT000005"
patient_subject = "51SY000110"

# ============================================================
# LOAD SIGNALS
# ============================================================

def load_signal(subject_id):

    file = (
        ROOT
        / subject_id
        / "denoised_signal.csv"
    )

    df = pd.read_csv(file, encoding='utf-8-sig')

    signal_col = None

    for c in df.columns:
        if "Pressure" in c or "Reconstructed" in c:
            signal_col = c
            break

    if signal_col is None:
        signal_col = df.columns[-1]

    signal = df[signal_col].values

    return signal


healthy_signal = load_signal(healthy_subject)
patient_signal = load_signal(patient_subject)

# 截取同长度
N = 1200

healthy_signal = healthy_signal[:N]
patient_signal = patient_signal[:N]

# ============================================================
# RECURRENCE PLOT
# ============================================================

rp = RecurrencePlot(
    dimension=3,
    time_delay=2,
    threshold="point",
    percentage=5,
)

healthy_rp = rp.fit_transform(
    healthy_signal.reshape(1, -1)
)[0]

patient_rp = rp.fit_transform(
    patient_signal.reshape(1, -1)
)[0]

# ============================================================
# PLOT
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(
    healthy_rp,
    cmap="binary",
    origin="lower"
)

axes[0].set_title(
    f"Healthy recurrence plot\n{healthy_subject}",
    fontsize=16
)

axes[1].imshow(
    patient_rp,
    cmap="binary",
    origin="lower"
)

axes[1].set_title(
    f"Patient recurrence plot\n{patient_subject}",
    fontsize=16
)

for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()

plt.savefig(
    OUT_DIR / "recurrence_plot_examples.png",
    dpi=600,
    bbox_inches="tight"
)

# plt.close()
plt.show()

print("Recurrence plots saved.")
```
